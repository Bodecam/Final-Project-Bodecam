Gráfico gerado para mostrar a média salarial de cada estado do Brasil.

```python
plt.figure(figsize=(15, 10))

sns.barplot(data = df_finalproject,
              x = 'Media Salarial',
              hue = 'UF')

plt.title('Média Salarial')
plt.xlabel('Valor Bruto')
plt.ylabel('Estados')

plt.xticks(rotation=90)

plt.legend(title = 'UF')

plt.show()
```
